```python
data['age'] = pd.DatetimeIndex(data['dt_customer']).year - data['year_birth'].astype(int)
data['age']
```

```python
data['age'] = pd.datetime.now().year - data['year_birth'].astype(int)
data['age']
```

```python
def age_group(age):
  if age < 21:
    age_group = 'teenager'
 
  elif age < 50:
    age_group = 'worker'
  
  else:
    age_group = 'boomer'
  return age_group
  
data['age_group'] = data['age'].apply(lambda x: age_group(x))
data['age_group'].value_counts()
```

```python
data[data['age_group'] == 'teenager'][['age', 'age_group']].reset_index()
```

```python
data = data.sort_values(by=['recency', 'income'], ascending=[True, False]).reset_index()
data
```

```python
def marital_status_category(marital_status):
  if marital_status == 'Married':
    marital_status = 'Together'
 
  elif marital_status == 'Single' or marital_status == 'Divorced' or marital_status == 'Widow' or marital_status == 'YOLO' or marital_status == 'Absurd':
    marital_status = 'Alone'
  
  return marital_status

data['marital_status'] = data['marital_status'].apply(lambda x: marital_status_category(x))
data['marital_status'].value_counts()
```
